```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "/content/estrapathways.txt"
data = []
current_concentration = None
with open(file_path, "r") as file:
  for line in file:
    line = line.strip()
    if not line:
      continue
    parts = line.split()
    try:
      concentration = float(parts[0])
      current_concentration = concentration
    except ValueError:
      hallmark = parts[0]
      score = float(parts[1])
      data.append([current_concentration, hallmark, score])

df = pd.DataFrame(data, columns=["Concentration", "Hallmark", "Score"])
df_pivot = df.pivot(index="Concentration", columns="Hallmark", values="Score")
df = pd.DataFrame(data, columns=["Concentration", "Hallmark", "Score"])
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Concentration", y="Score", hue="Hallmark", marker="o")
plt.axhline(0, color="black", linestyle="--", linewidth=1)
xticks = [0.0001, 0.001, 0.01, 0.1, 1, 10]
xtick_labels = [f"{x} \u03bcM" for x in xticks]

plt.xticks(xticks, xtick_labels)
plt.xscale("log")  # Log scale for better visualization
plt.xlabel("Concentration")
plt.ylabel("Score")
plt.title("Hallmark Pathway Scores Across Estradiol upregulated Concentrations (0.0001-10 nm)")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage

de_results = pd.read_csv("/content/fileFORbiostatsquidALLSTAT.csv", index_col=0)

#norm_counts = pd.read_csv("normalized_counts.csv", index_col=0, skiprows=2, header=3)


de_results.index = de_results.index.str.strip()
norm_counts.index = norm_counts.index.str.strip()
threshold_log2FC = 1.5
threshold_baseMean = 20

filtered_genes = de_results[
    (de_results["baseMean"] > threshold_baseMean) &
    (abs(de_results["log2FoldChange"]) > threshold_log2FC)
].index
valid_genes = filtered_genes.intersection(norm_counts.index)
filtered_counts = norm_counts.loc[valid_genes]
missing_genes = set(filtered_genes) - set(valid_genes)
if missing_genes:
    print(f"Warning: {len(missing_genes)} genes were not found in norm_counts and were removed.")
filtered_counts_zscore = filtered_counts.apply(lambda x: (x - x.mean()) / x.std(), axis=1)
row_linkage = linkage(filtered_counts_zscore, method="average", metric="euclidean")
col_linkage = linkage(filtered_counts_zscore.T, method="average", metric="euclidean")
sns.clustermap(
    filtered_counts_zscore,
    row_cluster=True, col_cluster=True,
    row_linkage=row_linkage, col_linkage=col_linkage,
    cmap="coolwarm", linewidths=0.5, figsize=(12, 8)
)

plt.title("Differential Expression Heatmap (Z-score)")
plt.show()
```
